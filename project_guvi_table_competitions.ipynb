```python
import requests

def get_comps_data(api_key, access_level='trial', version='v3', language='en', format_type='json'):
    url = f"https://api.sportradar.com/tennis/{access_level}/{version}/{language}/competitions.{format_type}?api_key={api_key}"
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None


api_key = "4KfXBNmeB6sSKb5p5RFdmZIsGbrKBpksS1mTRAej"  # Sportradar API key
data = get_comps_data(api_key)
    
if data:
    print(data) 
```
