```python
class competitors_ranking:
    def __init__(self,ta):
      self.table=ta
      self.doubles_rankings=[]

    def ranking_table(self,doubles_rankings):
        if self.table in ["rank","movement","points","competitions_played"]:
            for num in range(len(doubles_rankings)):
                 for length in range(len(doubles_rankings[num]['competitor_rankings'])):
                      keyvalue= doubles_rankings[num]['competitor_rankings'][length][self.table]
                      self.doubles_rankings.append(keyvalue)
        return self.doubles_rankings  

        
    def competitors_table(self,doubles_rankings):         
        if self.table in ["id","name","country","country_code","abbreviation"]:
              for num in range(len(doubles_rankings)):
                   for length in range(len(doubles_rankings[num]['competitor_rankings'])):
                        if self.table in doubles_rankings[num]['competitor_rankings'][length]['competitor']:
                            value= doubles_rankings[num]['competitor_rankings'][length]['competitor'][self.table]
                            self.doubles_rankings.append(value)
                        else:
                            if self.table in ["country_code"]:
                                self.doubles_rankings.append("NEU")
        return self.doubles_rankings                  
```
