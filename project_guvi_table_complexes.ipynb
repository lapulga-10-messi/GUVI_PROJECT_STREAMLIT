```python
class Complexes:
    def __init__(self, t):
        self.table = t
        self.complexes_table = []  

    def req_data(self, complexes_table):
        if self.table in ["id", "name", "city_name", "country_name", "country_code", "timezone"]: 
            for everyitem in complexes_table:
                if 'venues' in everyitem: 
                    for i in range(len(everyitem['venues'])):
                        value = everyitem['venues'][i][self.table]  
                        self.complexes_table.append(value)
        return self.complexes_table
```
